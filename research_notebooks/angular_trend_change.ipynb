# Detecting trend changes in angle



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sp_data_util as sp_data
```

```python
N = 5
x = np.linspace(0,10)
y = sp_data.ySampleNoisy(x,sp_data.kLIN, [1,1,0],0,.5,N)
[plt.plot(x,y_i,'.') for y_i in y]
plt.show()
```

```python
K = 5
N = 50
offsets = np.random.rand(K)*10
widths = np.random.rand(K)*5+2
x = [(np.random.rand(N)-.5)*w + o for o,w in zip(offsets,widths)]
y = [sp_data.ySampleNoisy(x_i,sp_data.kLIN, [1,5,o],0,.5,1)[0] + o for o,x_i in zip(offsets,x)]

[plt.plot(x_i,y_i,'.') for x_i,y_i in zip(x,y)]
plt.plot(offsets,offsets)
plt.show()
```

```python
c = [j for i in ['a','b','c','d','e'] for j in [i]*N]
x_dat = np.reshape(np.asarray(x),[N*K,1])
y_dat = np.reshape(np.asarray(y),[N*K,1])
plt.scatter(x_dat[:N],y_dat[:N],c=c)
plt.show()
```
